```python
# ============================================================
# DistilBERT Grid Search (LR, Epochs, Batch, Max Length) + Final Test
# ============================================================

!pip install -q transformers scikit-learn matplotlib seaborn

# --- Colab Drive ---
from google.colab import drive
drive.mount('/content/drive')

# --- Imports ---
import os, random, re, numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import label_binarize

import torch
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import (
    DistilBertTokenizerFast, DistilBertForSequenceClassification,
    get_linear_schedule_with_warmup
)

# --- Reproducibility ---
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# --- Load Data ---
CSV_PATH = '/content/drive/My Drive/annotatted_dataset.csv'
TEXT_COL = 'Base_Reviews'
LABEL_COL = 'category'

df = pd.read_csv(CSV_PATH)

# --- Map labels to stable IDs and keep class names ---
# If category is text, create codes; if already numeric, still create canonical names list
cat = df[LABEL_COL].astype('category')
class_labels = list(cat.cat.categories)
name_to_idx = {n:i for i,n in enumerate(class_labels)}
df['label_id'] = cat.cat.codes.values
n_classes = len(class_labels)

print("Classes:")
for i, n in enumerate(class_labels): print(i, '->', n)

# --- Stratified split: Train/Val/Test = 80/10/10 ---
idx_all = np.arange(len(df))
y_all = df['label_id'].values

# First: hold-out Test 10%
trainval_idx, test_idx = train_test_split(
    idx_all, test_size=0.10, random_state=SEED, stratify=y_all
)
# Then: Val 10% of total (i.e., 1/9 of trainval)
y_trainval = y_all[trainval_idx]
train_idx, val_idx = train_test_split(
    trainval_idx, test_size=0.1111, random_state=SEED, stratify=y_trainval
)

df_train = df.iloc[train_idx].reset_index(drop=True)
df_val   = df.iloc[val_idx].reset_index(drop=True)
df_test  = df.iloc[test_idx].reset_index(drop=True)

print(f"Split sizes -> Train: {len(df_train)}, Val: {len(df_val)}, Test: {len(df_test)}")

# --- Tokenizer ---
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')

def tokenize_df(dframe, max_len):
    return tokenizer(
        dframe[TEXT_COL].astype(str).tolist(),
        padding='max_length', truncation=True, max_length=max_len,
        return_tensors='pt', return_attention_mask=True
    )

# --- Dataset wrapper ---
class TextDS(Dataset):
    def __init__(self, toks, labels):
        self.ids = toks['input_ids']
        self.msk = toks['attention_mask']
        self.y   = torch.tensor(labels, dtype=torch.long)
    def __len__(self): return len(self.y)
    def __getitem__(self, i):
        return {'input_ids': self.ids[i],
                'attention_mask': self.msk[i],
                'labels': self.y[i]}

# --- Train/Eval utilities ---
def train_one_run(train_loader, val_loader, epochs, lr, weight_decay=0.01):
    model = DistilBertForSequenceClassification.from_pretrained(
        'distilbert-base-uncased', num_labels=n_classes
    ).to(device)

    optimizer = AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    total_steps = len(train_loader) * epochs
    scheduler = get_linear_schedule_with_warmup(
        optimizer, num_warmup_steps=0, num_training_steps=total_steps
    )

    best_val_acc = -1.0
    best_state = None

    for ep in range(epochs):
        # --- Train ---
        model.train()
        for batch in train_loader:
            optimizer.zero_grad()
            inputs = batch['input_ids'].to(device)
            masks  = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)
            outputs = model(inputs, attention_mask=masks, labels=labels)
            loss = outputs.loss
            loss.backward()
            optimizer.step(); scheduler.step()

        # --- Validate ---
        model.eval()
        val_correct, val_seen = 0, 0
        with torch.no_grad():
            for batch in val_loader:
                inputs = batch['input_ids'].to(device)
                masks  = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)
                logits = model(inputs, attention_mask=masks).logits
                preds = torch.argmax(logits, dim=1)
                val_correct += (preds == labels).sum().item()
                val_seen    += labels.size(0)
        val_acc = val_correct / max(1, val_seen)

        if val_acc > best_val_acc + 1e-6:
            best_val_acc = val_acc
            best_state = {k: v.detach().cpu().clone() for k, v in model.state_dict().items()}

        print(f"  Epoch {ep+1}/{epochs}  |  Val Acc: {val_acc:.4f}")

    # load best weights (by val acc over these epochs)
    if best_state is not None:
        model.load_state_dict({k: v.to(device) for k, v in best_state.items()})
    return model, best_val_acc

def evaluate_on_loader(model, data_loader):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for batch in data_loader:
            inputs = batch['input_ids'].to(device)
            masks  = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)
            logits = model(inputs, attention_mask=masks).logits
            preds = torch.argmax(logits, dim=1)
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())
    y_true = np.array(y_true); y_pred = np.array(y_pred)
    acc = (y_true == y_pred).mean() if len(y_true) else 0.0
    return acc, y_true, y_pred

def plot_confusion(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(9,7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_labels, yticklabels=class_labels)
    plt.title(title); plt.xlabel('Predicted'); plt.ylabel('True')
    plt.tight_layout(); plt.show()

# =========================
# GRID (edit these lists)
# =========================
LR_LIST       = [1e-5, 2e-5, 3e-5]
BATCH_LIST    = [8, 16]
EPOCHS_LIST   = [6, 8]
MAXLEN_LIST   = [128]   # Increase if VRAM allows; try [128] first if OOM

results = []

# --- Grid Search ---
for max_len in MAXLEN_LIST:
    # Tokenize once per max_len (for speed)
    tok_train = tokenize_df(df_train, max_len)
    tok_val   = tokenize_df(df_val,   max_len)

    y_train = df_train['label_id'].values
    y_val   = df_val['label_id'].values

    for batch_size in BATCH_LIST:
        train_loader = DataLoader(TextDS(tok_train, y_train), batch_size=batch_size, shuffle=True)
        val_loader   = DataLoader(TextDS(tok_val,   y_val),   batch_size=batch_size)

        for lr in LR_LIST:
            for epochs in EPOCHS_LIST:
                print(f"\n=== Combo: max_len={max_len}, batch={batch_size}, lr={lr}, epochs={epochs} ===")
                model_gs, best_val_acc = train_one_run(train_loader, val_loader, epochs=epochs, lr=lr)
                results.append({
                    'max_len': max_len,
                    'batch_size': batch_size,
                    'lr': lr,
                    'epochs': epochs,
                    'best_val_acc': best_val_acc
                })

# --- Show grid results ---
res_df = pd.DataFrame(results).sort_values(by='best_val_acc', ascending=False).reset_index(drop=True)
print("\n===== Grid Search Results (sorted by best validation accuracy) =====")
print(res_df.head(20))

# --- Pick best combo ---
best = res_df.iloc[0].to_dict()
print("\nBest combo:", best)

# =========================
# Final training on Train+Val with best hyperparameters, then Test
# =========================
best_len   = int(best['max_len'])
best_bs    = int(best['batch_size'])
best_lr    = float(best['lr'])
best_epoch = int(best['epochs'])

# Tokenize train+val and test with best max_len
tok_trainval = tokenize_df(pd.concat([df_train, df_val], ignore_index=True), best_len)
y_trainval   = np.concatenate([df_train['label_id'].values, df_val['label_id'].values])

tok_test = tokenize_df(df_test, best_len)
y_test   = df_test['label_id'].values

trainval_loader = DataLoader(TextDS(tok_trainval, y_trainval), batch_size=best_bs, shuffle=True)
test_loader     = DataLoader(TextDS(tok_test,     y_test),     batch_size=best_bs)

# Fresh model & optimizer
final_model = DistilBertForSequenceClassification.from_pretrained(
    'distilbert-base-uncased', num_labels=n_classes
).to(device)
final_opt = AdamW(final_model.parameters(), lr=best_lr, weight_decay=0.01)
final_steps = len(trainval_loader) * best_epoch
final_sched = get_linear_schedule_with_warmup(final_opt, num_warmup_steps=0, num_training_steps=final_steps)

# Train final model
for ep in range(best_epoch):
    final_model.train()
    for batch in trainval_loader:
        final_opt.zero_grad()
        inputs = batch['input_ids'].to(device)
        masks  = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)
        out = final_model(inputs, attention_mask=masks, labels=labels)
        loss = out.loss
        loss.backward()
        final_opt.step(); final_sched.step()
    print(f"[FINAL] Epoch {ep+1}/{best_epoch} done.")

# Evaluate on Test
test_acc, y_true_test, y_pred_test = evaluate_on_loader(final_model, test_loader)
print(f"\n=== FINAL Test Accuracy (best combo, trained on Train+Val) === {test_acc:.4f}\n")
print("Classification Report (Test):")
print(classification_report(y_true_test, y_pred_test, target_names=class_labels))

# Plot Confusion Matrix (Test)
plot_confusion(y_true_test, y_pred_test, "Confusion Matrix (Test)")

# Save best model
SAVE_DIR = '/content/drive/My Drive/distilbert_best_grid'
os.makedirs(SAVE_DIR, exist_ok=True)
final_model.save_pretrained(SAVE_DIR)
tokenizer.save_pretrained(SAVE_DIR)
print(f"\nSaved best model + tokenizer to: {SAVE_DIR}")
print("Best hyperparameters:", best)
```

```python
# ============================
# Finalize with RoBERTa (base)
# Combo: max_len=128, batch=8, lr=3e-5, epochs=6
# ============================
from transformers import RobertaTokenizerFast, RobertaForSequenceClassification, get_linear_schedule_with_warmup
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
import torch
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# --- Hyperparameters (your chosen combo) ---
best_len   = 128
best_bs    = 8
best_lr    = 3e-5
best_epoch = 6

# --- Tokenizer (RoBERTa) ---
roberta_tokenizer = RobertaTokenizerFast.from_pretrained('roberta-base')

def tokenize_df_roberta(dframe, max_len):
    return roberta_tokenizer(
        dframe['Base_Reviews'].astype(str).tolist(),
        padding='max_length', truncation=True, max_length=max_len,
        return_tensors='pt', return_attention_mask=True
    )

# --- Dataset wrapper (same as before) ---
class TextDS(Dataset):
    def __init__(self, toks, labels):
        self.ids = toks['input_ids']
        self.msk = toks['attention_mask']
        self.y   = torch.tensor(labels, dtype=torch.long)
    def __len__(self): return len(self.y)
    def __getitem__(self, i):
        return {
            'input_ids': self.ids[i],
            'attention_mask': self.msk[i],
            'labels': self.y[i]
        }

# --- Tokenize Train+Val and Test with RoBERTa tokenizer ---
tok_trainval = tokenize_df_roberta(pd.concat([df_train, df_val], ignore_index=True), best_len)
y_trainval   = np.concatenate([df_train['label_id'].values, df_val['label_id'].values])

tok_test = tokenize_df_roberta(df_test, best_len)
y_test   = df_test['label_id'].values

trainval_loader = DataLoader(TextDS(tok_trainval, y_trainval), batch_size=best_bs, shuffle=True)
test_loader     = DataLoader(TextDS(tok_test,     y_test),     batch_size=best_bs)

# --- Model (RoBERTa) + Optimizer + Scheduler ---
roberta_model = RobertaForSequenceClassification.from_pretrained(
    'roberta-base', num_labels=n_classes
).to(device)

optimizer = AdamW(roberta_model.parameters(), lr=best_lr, weight_decay=0.01)

total_steps = len(trainval_loader) * best_epoch
scheduler = get_linear_schedule_with_warmup(
    optimizer, num_warmup_steps=0, num_training_steps=total_steps
)

# --- Train on Train+Val ---
for ep in range(best_epoch):
    roberta_model.train()
    ep_loss, correct, seen = 0.0, 0, 0
    for batch in trainval_loader:
        optimizer.zero_grad()
        ids = batch['input_ids'].to(device)
        msk = batch['attention_mask'].to(device)
        lbl = batch['labels'].to(device)
        out = roberta_model(ids, attention_mask=msk, labels=lbl)
        loss = out.loss
        loss.backward()
        optimizer.step(); scheduler.step()
        ep_loss += loss.item()
        pred = torch.argmax(out.logits, dim=1)
        correct += (pred == lbl).sum().item(); seen += lbl.size(0)
    print(f"[RoBERTa FINAL] Epoch {ep+1}/{best_epoch}  loss={ep_loss/len(trainval_loader):.4f}  acc={correct/seen:.4f}")

# --- Evaluate on TEST ---
roberta_model.eval()
y_true, y_pred = [], []
with torch.no_grad():
    for batch in test_loader:
        ids = batch['input_ids'].to(device)
        msk = batch['attention_mask'].to(device)
        lbl = batch['labels'].to(device)
        logits = roberta_model(ids, attention_mask=msk).logits
        pred = torch.argmax(logits, dim=1)
        y_true.extend(lbl.cpu().numpy()); y_pred.extend(pred.cpu().numpy())

y_true = np.array(y_true); y_pred = np.array(y_pred)
test_acc = (y_true == y_pred).mean()
print(f"\n=== RoBERTa FINAL Test Accuracy (len=128, batch=8, lr=3e-5, epochs=6) === {test_acc:.4f}\n")

print("Classification Report (Test):")
print(classification_report(y_true, y_pred, target_names=class_labels))

# --- Confusion Matrix (Test) ---
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(9,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix (Test) — RoBERTa')
plt.xlabel('Predicted'); plt.ylabel('True')
plt.tight_layout(); plt.show()
```
